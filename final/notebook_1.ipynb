```python
inflation = pd.read_csv('data/inflation.csv')
infl_sorted = inflation.set_index('Год').sort_values(by='Год')
```

```python
x = infl_sorted['Всего']
```

```python
res = pd.concat([df, x], axis=1, join='inner')
res
```

```python
res['coeffs'] = 1 + (res['Всего'] /100) 
r_sal = list(res['coeffs']) 
```

```python
import numpy as np
real_salaries_coeffs = []
real_salaries_coeffs.append(1)
x = 1
for i in range(1, len(r_sal)):
    x = x / r_sal[i]
    real_salaries_coeffs.append(x)

real_salaries_coeffs = np.array(real_salaries_coeffs)
real_salaries_coeffs
```

```python
real_salary = res['сельское хозяйство, охота и лесное хозяйство'] * real_salaries_coeffs
real_salary.plot(kind='bar')
```
